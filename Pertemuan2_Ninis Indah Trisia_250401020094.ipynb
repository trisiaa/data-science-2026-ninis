## Pertanyaan Latihan

### Q1 Berapa jumlah baris dan kolom pada dataset Titanic? Kolom apa saja yang tersedia?

```python
print(df.shape)
print(df.columns)
```

Jadi jumlah baris pada dataset Titanic adalah 891 dan jumlah kolomnya adalah 12.

### Q2 Kolom mana saja yang memiliki missing values? Berapa jumlah dan proporsinya (%) per kolom?

```python
missing = df.isnull().sum()
persentase = (missing / len(df)) * 100

hasil_missing = pd.DataFrame({
    'Jumlah Missing': missing,
    'Persentase (%)': persentase
})

print(hasil_missing)
```

Kolom yang memiliki missing value adalah sebagai berikut:

*   Kolom age dengan jumlah missing 177 dan
persentase 19.865320%
*   Kolom cabin dengan jumlah missing 687 dan
persentase 77.104377%
*   Kolom embarked dengan jumlah missing 2 dan
persentase 0.224467%





### Q3 Berapa persen penumpang yang selamat (Survived=1)? Gunakan operasi NumPy atau Pandas.

```python
persen_selamat = df['Survived'].mean() * 100
print(f"Persentase penumpang selamat: {persen_selamat:.2f}%")
```

Jadi persentase penumpang selamat ada 38.38%

### Q4 Filter penumpang wanita dari kelas 1 (Pclass=1). Berapa jumlahnya dan berapa rata-rata usianya?

```python
wanita_kelas1 = df[(df['Sex'] == 'female') & (df['Pclass'] == 1)]

jumlah = len(wanita_kelas1)
rata_usia = wanita_kelas1['Age'].mean()

print("Jumlah:", jumlah)
print("Rata-rata usia:", rata_usia)
```

Berdasarkan filter penumpang wanita dari kelas 1 (Pclass=1), jumlah penumpang wanita adalah 94 dengan rata-rata usia 34.61176470588235

### Q5 Gunakan groupby untuk menghitung tingkat keselamatan (mean Survived) per kelas penumpang. Kelas mana yang paling tinggi?

```python
keselamatan = df.groupby('Pclass')['Survived'].mean()
print(keselamatan)
```

Jadi hasil groupby, tingkat keselamatan tertinggi ada pada (Pclass) 1 dengan tingkat keselamatan 0.629630

### Q6 Buat analisis tambahan berdasarkan variabel lain yang Anda anggap menarik. Sertakan interpretasi singkat.

```python
fare_kelas = df.groupby('Pclass')['Fare'].mean()
print(fare_kelas)
```

Analisis tambahan yang saya sertakan adalah menghitung rata-rata tarif tike (Fare) berdasarkan kelas penumpang (Pclass). Hasil perhitungan menunjukkan bahwa kelas 1 memiliki rata-rata tarif tertinggi yaitu 84.154687%
